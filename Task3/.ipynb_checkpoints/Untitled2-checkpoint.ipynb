```python
import pandas as pd

df = pd.read_excel("transformed6.xlsx")
df = df.drop(['Vehicle 1 Trim', 'Engine Size', 'Engine Block', 'Auto In The Market: New', 'Auto In The Market: Used', 'Model 1', 'Engine Cylinders', 'Drive Type', 'Fuel Type', 'Make 1', 'Unnamed: 0'], axis=1)
```

```python
df.columns
```

```python
for col in df.columns:
    print(col)
    l1 = []
    l2 = []
    i = 0
    j = 0
    q = 1
    x = len(df[col].unique())
    if(x<=4):
        continue
    else:
        for i in range(0,len(df)):
            x = df.loc[i,col]
            if(j < 4):
                if(x in l1):
                    df.loc[i,col] = l2[l1.index(x)]
                else:    
                    l1.append(x)
                    l2.append(q)
                    df.loc[i,col] = q
                    q = q + 1
                    j = j + 1
            else:
                if(x == l1[0]):
                    df.loc[i,col] = l2[0]
                elif(x == l1[1]):
                    df.loc[i,col] = l2[1]
                elif(x == l1[2]):
                    df.loc[i,col] = l2[2]    
                else:
                    df.loc[i,col] = 4
```
